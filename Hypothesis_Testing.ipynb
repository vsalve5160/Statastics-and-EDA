Bombay hospitality Ltd. operates a franchise model for producing exotic Norwegian dinners throughout New England. The operating cost for a franchise in a week (W) is given by the equation W = $1,000 + $5X, where X represents the number of units produced in a week. Recent feedback from restaurant owners suggests that this cost model may no longer be accurate, as their observed weekly operating costs are higher.
Objective:
To investigate the restaurant owners' claim about the increase in weekly operating costs using hypothesis testing.

```python
import math
from scipy import stats   #
```

```python
# Given values
xbar = 3050              # sample mean
mu0 = 1000 + 5*600       # theoretical mean = 4000
sigma = 5 * 25           # population std dev = 125
n = 25                   # sample size
```

```python
# Standard error and Z-statistic
se = sigma / math.sqrt(n)
z_stat = (xbar - mu0) / se
se
```

Hypotheses statement:

Null Hypothesis (Hâ‚€):
ðœ‡=4000

(No increase in cost; model is accurate)

Alternative Hypothesis (Hâ‚):

ðœ‡>4000
(Costs are higher than what the model suggests)

```python
# Critical value for one-tailed test
alpha = 0.05
z_crit = stats.norm.ppf(1 - alpha)   # one-tailed (right tail)
```

```python
# Decision
print("\nTest Statistic (Z):", z_stat)
print("Critical Value (Z_crit):", z_crit)
```

```python
if z_stat > z_crit:
    print("Decision: Reject H0 â†’ Costs are significantly higher.")
else:
    print("Decision: Fail to Reject H0 â†’ No strong evidence of higher costs.")
```
