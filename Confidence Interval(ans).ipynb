# Confidence Intervel

1  Import Libraries

```python
# Scipy for Stats Function Data  & Pandas and Numpy for data Manupulation

from scipy import stats
import pandas as pd
import numpy as np
```

2Import Data

```python
glaxo_df=pd.read_csv('glaxo_df.csv')
glaxo_df.head()
```

```python
#95% Confidence interval Confidence interval=((n,loc/Mean,SD/scale)
glaxo_df_ci = stats.norm.interval(0.95,
loc = glaxo_df.gain.mean(),
scale = glaxo_df.gain.std())
print( 'Gain at 95% confidence interval is:', np.round(glaxo_df_ci, 2))
```

```python
beml_df_ci=pd.read_csv('beml_df.csv')
beml_df.head()
beml_df_ci = stats. norm.interval(0.95,
loc=beml_df.gain.mean(),
scale=beml_df.gain.std())
print( 'Gain at 95% confidence interval is:', np.round(beml_df_ci, 4))
```

```python
#Gain at 99%,90% Confidence Interval

glaxo_df_ci = stats.norm.interval(0.99,
loc=glaxo_df.gain.mean(),
scale=glaxo_df.gain.std())
print( 'Gain at 99% confidence interval is:', np.round(glaxo_df_ci, 4))
```

```python
beml_df_ci=pd.read_csv('beml_df.csv')
beml_df.head()
beml_df_ci = stats. norm.interval(0.90,
loc=beml_df.gain.mean(),
scale=beml_df.gain.std())
print( 'Gain at 95% confidence interval is:', np.round(beml_df_ci, 4))
```

```python
#Probability of making 2% gain or higher in Glaxo
GainMorethan2 =1 - stats.norm.cdf(0.02,
loc=glaxo_df.gain.mean(),
scale=glaxo_df.gain.std())
print("There is chance that the inverster invest in glaxo will be gaining   2% gain or higher",round(GainMorethan2*100,2))
```

```python
#Probability of making 2% loss or higher in Glaxo
GainMorethan2 =stats.norm.cdf(0.02,
loc=glaxo_df.gain.mean(),
scale=glaxo_df.gain.std())
print("There is chance that the inverster invest in glaxo will be lossing 2% gain or higher",
      round(GainMorethan2*100,2),"%")
```
