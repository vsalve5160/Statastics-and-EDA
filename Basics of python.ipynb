# prime number

```python
num = int(input("Enter a number: "))

if num <= 1:
    print("Not a prime number")  # 0, 1, and negative numbers are not prime
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not a prime number")
            break
    else:
        print("Prime number")
```

# Product of Random Numbers


```python
import random

num1= random.randint(1,10)
num2= random.randint(1,10)

print(f"What is the product of {num1} x {num2}")
your_ans=int(input("your_ans:"))

Ans= num1* num2

if your_ans==Ans:
    print("Correct")
else:
    print(f"Incorrect, The correct answer is {Ans}.")
```

# Squares of Even/Odd Numbers

```python


for num in range(100, 201):
    if num % 2 == 0:
        print(f"EVEN: {num}^2 = {num ** 2}")
    else:
        print(f"ODD : {num}^2 = {num ** 2}")
```

# Word counter

```python
input_text="This is a sample text. This text will be used to demonstrate the word counter."

input_text.replace("."," ")
input_text.lower()

words=input_text.split()

word_count={}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1
        
for word,count in word_count.items():
    print(f" '{word}':{count}")
```

# Check for Palindrome

```python
import string

user_input = input("Enter a word or sentence: ")

def is_palindrome(text):
    
     return text == text[::-1]

# Call the function and print the result
if is_palindrome(user_input):
    print("✅ It is a palindrome!")
else:
    print("❌ It is not a palindrome.")
```
