```python
import pandas as pd
import scipy
import numpy as np
from scipy.stats import t, norm
```

```python
data=[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
len(data)
```

```python
x=np.mean(data)
```

```python
s = np.std(data, ddof=1)#dd0f=1 because it divides by n – 1 (used for sample standard deviation)
```

```python
σ = 0.2  # Known population std dev
α = 0.01  # For 99% confidence
df = len(data) - 1

```

```python
#a. Build 99% Confidence Interval Using Sample Standard Deviation
```

# T-Test

```python
#Sample size is small (n < 30)
#Population standard deviation is not known
#We are using sample standard deviation instead
```

```python
t_score=t.ppf(1 - α/2, df)
print (t_score)
```

```python
ME_t = t_score * (s / np.sqrt(df)) # margin of error ,s= std of sample
print(ME_t)
```

```python
CI_t = (x - ME_t, x + ME_t) # confidence interval
print("confidence intervel using sample std",CI_t)
```

# Z- Test

```python
#b. Build 99% Confidence Interval Using Known Population Standard Deviation
```

```python

σ = 0.2  # Known population std dev
```

```python
z_score = norm.ppf(1 - α/2)
```

```python
ME_z = z_score * (σ / np.sqrt(df)) #σ =std of population
```

```python
CI_z = (x - ME_z, x + ME_z)
print(" Confidence Interval Using Known Population Std",CI_z)
```
