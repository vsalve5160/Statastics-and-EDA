We do not use Label Encoding blindly on input columns with fewer unique values (like gender, binary yes/no), because:
Label encoding gives order to the categories (0 < 1 < 2),
which might be misleading.
For such columns, One-Hot Encoding or pd.get_dummies() is better.

but here For input columns that are categorical with many unique values (high cardinality), like:
occupation,
native_country,
workclass,

```python
cat_cols = df.select_dtypes(include='object').columns.drop('income')
```

```python
high_cardinality = [col for col in cat_cols if df[col].nunique() > 5] #filters only columns with many unique values.
```

```python
# Apply Label Encoding
le = LabelEncoder()
for col in high_cardinality:
    df[col] = le.fit_transform(df[col])
```
